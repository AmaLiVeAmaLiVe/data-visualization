D√≠v√°m se na informace o datech ze datafram≈Ø a tak√© na poƒçet chybƒõj√≠c√≠ch dat

```python
def dataframe_info(dataframe):
    dataframe.info()
    dataframe.isnull().sum(axis=0)
    print(f'Pocet radku v danem dataframu: {len(dataframe)} \n\n')

print('Dataframe Intakes')
dataframe_info(intakes)
print('Dataframe Outcomes')
dataframe_info(outcomes)
```

D√≠v√°m se na v≈°echny unik√°tn√≠ hodnoty pro v≈°echny sloupce datarfram≈Ø

```python
def unique_data(dataframe):
    for x in dataframe.columns:
        print('\n')
        print(x)
        print(dataframe[x].nunique())
        print(dataframe[x].unique())

unique_data(intakes)
unique_data(outcomes)
```

Nahrazuju chybej√≠c√≠ hodnoty typem np.NaN

```python
intakes = intakes.replace('Unknown', np.nan)
outcomes = outcomes.replace('Unknown', np.nan)    
```

P≈ôevod√≠m vhodn√© sloupce s kategorick√Ωmi p≈ô√≠znaky na typ category

```python
for x in ['Intake Type', 'Intake Condition', 'Animal Type', 'Sex upon Intake']:
    intakes[x] = intakes[x].astype('category')

for x in ['Outcome Type', 'Outcome Subtype', 'Animal Type', 'Sex upon Outcome']:
    outcomes[x] = outcomes[x].astype('category')
```

Pomoc√≠ for cyklu p≈ôevod√≠m hodnoty ve sloupci "Age upon Intake/Outcome" na ƒç√≠sla float, abych pak mohl p≈ôev√©st cel√Ω sloupec na numerick√Ω typ 

```python
def to_numeric(dataframe, column):
    index = 0
    for x in dataframe[column]:      
        if type(x) is str:
            separated_string = x.split(' ')
            if float(separated_string[0]) <= 0:
                dataframe.at[index, column] = np.nan
            else:
                if separated_string[-1] == 'year' or separated_string[-1] == 'years':
                    dataframe.at[index, column] = float(separated_string[0])
                elif separated_string[-1] == 'month' or separated_string[-1] == 'months':
                    dataframe.at[index, column] = round(float(separated_string[0]) / 12, 2)
                elif separated_string[-1] == 'week'or separated_string[-1] == 'weeks':
                    dataframe.at[index, column] = round(float(separated_string[0]) / 52, 2)
                else:
                    dataframe.at[index, column] = round(float(separated_string[0]) / 365, 2)
        index += 1

    dataframe[column] = dataframe[column].astype(float)  

to_numeric(intakes, 'Age upon Intake')
to_numeric(outcomes, 'Age upon Outcome')
```

P≈ôevod√≠m sloupec "DateTime" na typ "datetime64[ns]"

```python
intakes['DateTime'] = pd.to_datetime(intakes['DateTime'], format="%m/%d/%Y %I:%M:%S %p")
outcomes['DateTime'] = pd.to_datetime(outcomes['DateTime'], format="%m/%d/%Y %I:%M:%S %p")
```

Sma≈æu zbyteƒçn√© sloupce - MonthYeat z obou datafram≈Ø

```python
intakes = intakes.drop('MonthYear', axis=1)
outcomes = outcomes.drop('MonthYear', axis=1)
```

Odstranƒõn√≠ v≈°ech duplicitn√≠ch dat

```python
def duplicates(dataframe):
    print(f'Pocet duplikovanuych radku v danem dataframu: {dataframe.duplicated().sum()}')
    dataframe.drop_duplicates()

print('Intakes')
duplicates(intakes)

print('\nOutcomes')
duplicates(outcomes)
```
