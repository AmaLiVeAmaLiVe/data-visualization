# ≈òe≈°en√≠

## I. P≈ô√≠prava dat

Importuju pot≈ôebn√© knihovny

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

import seaborn as sns
import networkx as nx

import csv

%matplotlib inline
```

Naƒçit√°m data ze souboru a d√≠v√°m se na informace o nich

```python
df = pd.read_csv('d.csv')
display(df.head())
display(df.info())
```

Vid√≠m, ≈æe n√°≈° dataframe neobsahuje ≈æ√°dn√° nullov√° data. Tedy, m≈Ø≈æu pracovat s daty bez probl√©m≈Ø

Uprav√≠m ƒçasov√© √∫daje pomoc√≠ vhodn√© funkce

```python
def reformat_time(dataframe, column):
    for index, data in enumerate(df[column]):
        time_array = data.split(':')
        if int(time_array[0]) > 23:
            time_array[0] = str(int(time_array[0]) % 24)
            dataframe.at[index, column] = ':'.join(time_array)


reformat_time(df, 'depart_from')
reformat_time(df, 'arrive_to')
```

Naƒçit√°m data s lokacemi ka≈æd√© zast√°vky a vytvo≈ô√≠m p≈ô√≠slu≈°n√Ω slovn√≠k

```python
file_path = 'stops.txt'
stops_pos = {}

with open(file_path, 'r', encoding='utf-8') as file:
    file.readline()
    csv_reader = csv.reader(file, delimiter=',')
    
    for line in csv_reader:
        stops_pos[line[1]] = np.array([float(line[3]), float(line[2])])
```

## II. Vizualizace grafu

P√≠≈°u funkci pro vytvo≈ôen√≠ grafu 

```python
def construct_graph(transport_df):
    G = nx.DiGraph()
    stops = transport_df['stop_from_name'].unique()
    G.add_nodes_from(stops)

    for index, row in transport_df.iterrows():
        G.add_edge(row['stop_from_name'], row['stop_to_name'])

    return G
```

Vytvo≈ô√≠m dataset jen s tramvajov√Ωmi √∫daji a p≈ô√≠slu≈°n√Ω k nim graf

```python
tram_df = df[df['route_type'] == 0]
tram_graph = construct_graph(tram_df)
```

Dƒõl√°m dal≈°√≠ √∫pravy pro vykryslen√≠ dobr√©ho grafu

```python
#Z√≠sk√°n√≠ koneƒçn√Ωch zast√°vek pomoc√≠ neorientovan√°ho grafu
terminal_stops = [node for node in tram_graph.nodes if tram_graph.to_undirected().degree(node) == 1]

#Vytvo≈ôen√≠ slovn√≠ku s n√°zvy koneƒçn√Ωch zast√°vek
labels = {}
for name, coord in stops_pos.items():
    if name in terminal_stops:
        labels[name] = name

#Vytvo≈ôen√≠ slovn√≠ku obsauj√≠c√≠ho n√°zvy koneƒçn√Ωch zast√°vek s p≈ô√≠slu≈°n√Ωmi lokacemi
label_pos = {}
for name, coord in stops_pos.items():
    if name in terminal_stops:
        label_pos[name] = coord

#Upraven√≠ lokac√≠ pomoc√≠ servisu Google Maps, aby se p≈ôi vykreslov√°n√≠ grafu nep≈ôekr√Ωvaly se zast√°vkami
label_pos['Radlick√°'] = np.array([50.055938, 14.387796][::-1])
label_pos['B√≠l√° Hora'] = np.array([50.073180, 14.321092][::-1])
label_pos['Lehovec'] = np.array([50.110416, 14.547864][::-1])
label_pos['Divok√° ≈†√°rka'] = np.array([50.096737, 14.320055][::-1])
label_pos['N√°dra≈æ√≠ Podbaba'] = np.array([50.114853, 14.393539][::-1])
label_pos['N√°dra≈æ√≠ Hostiva≈ô'] = np.array([50.055457, 14.538338][::-1])
label_pos['S√≠dli≈°tƒõ Pet≈ôiny'] = np.array([50.083689, 14.338574][::-1])
label_pos['S√≠dli≈°tƒõ ≈òepy'] = np.array([50.062002, 14.295307][::-1])
label_pos['Spojovac√≠'] = np.array([50.090999, 14.498117][::-1])
label_pos['Spo≈ôilov'] = np.array([50.048505, 14.483220][::-1])
label_pos['√öst≈ôedn√≠ d√≠lny DP'] = np.array([50.072634, 14.532004][::-1])
label_pos['Vozovna Kobylisy'] = np.array([50.136734, 14.453187][::-1])
label_pos['Levsk√©ho'] = np.array([50.003454, 14.435848][::-1])
label_pos['Vozovna Pankr√°c'] = np.array([50.058829, 14.436994][::-1])
label_pos['Holynƒõ'] = np.array([50.025495, 14.357576][::-1])
label_pos['Zvona≈ôka'] = np.array([50.069272, 14.434940][::-1])
```

Vykreslov√°n√≠ sam√©ho grafu

```python
plt.figure(figsize=(10, 10))
plt.title('Tram Graph')

nx.draw(tram_graph, 
        pos = stops_pos, 
        node_size=20, 
        node_color='yellow',
        width = 0.5,
        arrows=False)
nx.draw_networkx_nodes(tram_graph, pos=stops_pos, nodelist=tram_graph.nodes(), node_color='yellow', node_size=20, edgecolors='black')
nx.draw_networkx_nodes(tram_graph, pos=stops_pos, nodelist=terminal_stops, node_color='red', node_size=60, edgecolors='black')
nx.draw_networkx_labels(tram_graph, pos=label_pos, labels=labels, font_size=8)

plt.show()
```

## III. Anal√Ωza m√≠r centralit grafu

P√≠≈°u vhodnou funkci pro vykreslov√°n√≠ m√≠r centralit grafu

```python
#Zdroj: https://aksakalli.github.io/2017/07/17/network-centrality-measures-and-their-visualization.html
def draw_centrality(graph, centrality, name):
    plt.figure(figsize=(15, 10))
    nodes = nx.draw_networkx_nodes(graph, 
                           pos=stops_pos, 
                           cmap=plt.cm.plasma, 
                           node_color=list(centrality.values()),
                           nodelist=centrality.keys(),
                           node_size = 20)
    edges = nx.draw_networkx_edges(graph, stops_pos, arrows=False)

    plt.title(name)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
```

P√≠≈°u vhodnou funkci pro anal√Ωzu m√≠r centralit grafu

```python
def analyze_centrality(graph_centrality):
    top_stops = lowest_stops = []
    count = 1
    
    #Se≈ôazen√≠ slovn√≠ku vzestupnƒõ
    graph_centrality = dict(sorted(graph_centrality.items(), key=lambda item: item[1]))

    print('Nejv√≠ce v√Ωznamn√© 3 z√°stvaky:')    
    top_stops = list(graph_centrality.items())[-3:][::-1]
    for stop in top_stops:
        print(f'{count}) {stop[0]}: {round(stop[1],6)}')
        count += 1
        
    count = 1
    print('\nNejm√©nƒõ v√Ωznamn√© 3 z√°stvaky:')
    lowest_stops = list(graph_centrality.items())[:3]
    for stop in lowest_stops:
        print(f'{count}) {stop[0]}: {round(stop[1],6)}')
        count += 1
```

### III(a). Degree centrality

```python
tram_degree_centrality = nx.degree_centrality(tram_graph)
analyze_centrality(tram_degree_centrality)
```

Nejv√≠ce v√Ωznamn√© z√°stvaky podle degree centrality jsou Malostransk√°, N√°mƒõst√≠ Brat≈ô√≠ Synk≈Ø, Otakarova, co n√°m ≈ôik√°, ≈æe tyto zast√°vky maj√≠ nejv√≠c spojen√≠ s jin√°my zast√°vkami. To dƒõl√° z nich osobit√© transportn√≠ uzly a p≈ôisoud√≠ jim v√Ωznam podle cel√©ho dopravn√≠ho uzlu. Nach√°z√≠ se v centru Prahy a tedy se pou≈æ√≠vaj√≠ nejen obyvateli Prahy, ale tak√© touristi.

Nejm√©nƒõ v√Ωznamn√© z√°stvaky podle degree centrality jsou Levsk√©ho, S√≠dli≈°tƒõ ≈òepy, Spojovac√≠. To n√°m ≈ô√≠k√°, ≈æe tyto zast√°vky maj√≠ nejm√©nƒõ spojen√≠ s jin√°my zast√°vkami. Zm√≠n√≠m, ≈æe tyto zast√Ωvky jsou koneƒçn√© zast√°vky tramvajov√Ωch linek. M≈Ø≈æu udƒõlat z√°vƒõr, ≈æe tyto zast√°vky jsou lok√°ln√≠ a se pou≈æivaj√≠ mnohem jen m√≠stn√≠mi lidmi.

```python
draw_centrality(tram_graph, tram_degree_centrality, 'Degree Centrality')
```

### III(b). Closeness centrality

```python
tram_closeness_centrality = nx.closeness_centrality(tram_graph)
analyze_centrality(tram_closeness_centrality)
```

Nejv√≠ce v√Ωznamn√© z√°stvaky podle closeness centrality jsou Malostransk√°, Staromƒõstsk√°, ƒåech≈Øv most. Tato centralita uk√°≈æe, ≈æe ty zast√°vky jsou nejd≈Øle≈æitƒõj≈°√≠ pro pasa≈æ√©ry, proto≈æe oni mohou se dostat do v≈°ech ostatn√≠ch zast√°vek co nejrychleji a nejkrat≈°i. Tak√© se nach√°z√≠ v centru mƒõsta.

Nejm√©nƒõ v√Ωznamn√© z√°stvaky podle closeness centrality jsou Levsk√©ho, N√°dra≈æ√≠ Hostiva≈ô, S√≠dli≈°tƒõ Mod≈ôany. To n√°m vyprav√≠, ≈æe pokud se chceme dostat do ostatn√≠ch zast√°vek odsud, to n√°m zabere spoustu ƒçasu a urƒçitƒõ je≈°tƒõ p≈ôestupy na tramvajov√© nebo jin√© transportn√≠ linky. Tyto zast√°vky jsou koneƒçn√©.

```python
draw_centrality(tram_graph, tram_closeness_centrality, 'Closeness Centrality')
```

### III(c). Betweenness centrality

```python
tram_betweenness_centrality = nx.betweenness_centrality(tram_graph)
analyze_centrality(tram_betweenness_centrality)
```

Nejv√≠ce v√Ωznamn√© z√°stvaky podle betweenness centrality jsou Malostransk√°, N√°rodn√≠ divadlo, ƒåech≈Øv most. Tato centrali n√°m ≈ô√≠k√°, jak ƒçasto se tyto zast√Ωvky pou≈æ√≠vaj√≠ jako p≈ôestupn√≠ podle cel√© tramvjov√© s√≠tƒõ. Nach√°z√≠ se v centru mƒõsta, aby mƒõli co nejv√≠c spojen√≠ s jin√Ωmi tramvaov√Ωmi linkami.

Nejm√©nƒõ v√Ωznamn√© z√°stvaky podle betweenness centrality jsou S√≠dli≈°tƒõ ≈òepy, Spojovac√≠, Vozovna Pankr√°c, co n√°m ≈ô√≠k√°, ≈æe se v≈Øbec nepou≈æ√≠vaj√≠ jako p≈ôestupn√≠ a jsou koneƒçn√© zast√°vky tramvajov√Ωch linek. Nemaj√≠ takov√Ω velk√Ω v√Ωznam pro celou tramvajovou sit'.

```python
draw_centrality(tram_graph, tram_betweenness_centrality, 'Betweenness Centrality')
```

## IV. Vlastn√≠ ot√°zky

##### 1) Jak√© jsou nejd≈Øle≈æitƒõj≈°√≠ zast√°vky v noƒçn√≠ch hodin√°ch?

Vytvo≈ô√≠m dataset obsahuj√≠c√≠ jen noƒçn√≠ tramvajov√© linky

##### 3) Jak√° je distribuce tramvajov√Ωch linek bƒõhem cel√©ho dne?

Vytvo≈ô√≠m slovn√≠k s poƒçtem tramvajov√Ωch j√≠zd podle ka≈æd√≠ho ƒçasu a z√°roven pole s barvami pro noƒçn√≠ a denn√≠ hodiny

```python
tram_df['depart_from'] = pd.to_datetime(tram_df['depart_from'], format='%H:%M:%S')
hours = {}
colors = []

for hour in range(0, 24):
    hours[hour] = tram_df[tram_df['depart_from'].dt.hour == hour].shape[0]
    if 6 < hour < 22:
        colors.append("#bd7ebe")
    else:
        colors.append("#8bd3c7")
```

```python
hour_series = pd.Series(hours)
plt.figure(figsize=(12,8))
for k,v in hour_series.items():
    plt.text(k, v, str(v), ha='center', va='bottom', fontsize=8)

sns.barplot(hour_series, edgecolor='black', palette=colors)

plt.title('Tram Network Traffic During the Day')
legend_labels = ['Daytime', 'Nighttime']
legend_colors = ["#bd7ebe", "#8bd3c7"]
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10) for color in legend_colors]
plt.legend(legend_elements, legend_labels, loc='upper right')
plt.xlabel('Hour')
plt.ylabel('Total')
plt.grid(linestyle='-.', alpha=0.5)

plt.show()
```

Data n√°m vypov√≠d√°, ≈æe celkem bƒõhem dne vyt√≠≈æenost tramvajov√Ωch linek v√Ω≈°e ne≈æ behem noci. Ale v 22 a 23 hodin noci jsou zast√°vky stejnƒõ vyt√≠≈æn√© jako v 21. V 5 a 6 r√°no tramvajov√© linky maj√≠ v√Ω≈°≈°√≠ vyt√≠≈æenost ne≈æ veƒçer. Je to kv≈Øli tomu, ≈æe zaƒç√≠n√° pracovn√≠ den. 7 a 8 jodin r√°no jsou hodiny dopravn√≠ ≈°piƒçky a proto maj√≠ nejv√Ω≈°≈°√≠ vyt√≠≈æenost bƒõhem cel√©ho dne. 16-18 hodin veƒçer se tak√© poƒç√≠taj√≠ jako hodiny dopravn√≠ ≈°piƒçky.
